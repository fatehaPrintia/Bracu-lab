```python
class book:
    def __init__(self,name):
        self.name=name
        self.genre='biography'
    def review(self):
        print('This book is just out of the world,mind-blowing!')
class fiction(book):
    def __init__(self,name,y):
        super().__init__(name)
        self.genre=y
    def review(self):
        print(self.name,'which is a',self.genre,'is just out of the world,mind-blowing!')
class nonfiction(book):
    def __init__(self,name):
        super().__init__(name)
    def review(self):
        print(self.name,'which is a',self.genre,'is just out of the world,mind-blowing!')
        
b1 = fiction('The Shining','Psychological horror')
b2 = nonfiction('A Beautiful Mind')
b1.review()
print('=========================')
b2.review()
print('=========================')

```

```python
class ParcelKoro:
    def __init__(self,name=None,weight=0):
        if name==None:
            self.name="No name set"
            self.product_weight=weight
            self.total=0
        else:
            self.name=name
            self.product_weight=weight
    def calculateFee(self,place=None):
        total=0
        if place==None and self.product_weight!=0:
            total=(self.product_weight * 20) + 50
        elif place==None and self.product_weight==0:
            total=0
        else:
            total=(self.product_weight * 20) + 100
        self.total=total
    def printDetails(self):
        print("Customer Name:", self.name) 
        print("Product Weight: ",self.product_weight)
        print("Total fee:",self.total)
print("**")
p1 = ParcelKoro()
p1.calculateFee()
p1.printDetails()
print("\n**")
p2 = ParcelKoro('Bob The Builder')
p2.calculateFee()
p2.printDetails()
print("----------------------------")
p2.product_weight = 15
p2.calculateFee()
p2.printDetails()
print("\n**")
p3 = ParcelKoro('Dora The Explorer', 10)
p3.calculateFee('Dhanmondi')
p3.printDetails()
```

```python
class Avenger:
    def __init__(self,name=None,power=None):
        if name==None and power==None:
            print('Super hero name and power need to be set')
        elif name!=None and power==None:
            self.__name=name
            print('power for',name,'needs to be set')
        elif name!=None and power!=None:
            self.__name=name
            self.__power=power
            print(name,'is ready to attack')
        else:
            self.__name=name
            self.__power=power
    def setName(self,name):
        self.__name=name
    def setPower(self,power):
        self.__power=power
    def getName(self,name):
        return self.__name
    def getPower(self,power):
        return self.__power
    def attack(self):
        print(self.__name+'*is*',self.__power+'ing')
tony=Avenger()
tony.setName("Iron Man")
tony.setPower("blast")
natasa=Avenger("Black Widow")
natasa.setPower("shoot")
bruce=Avenger("Hulk","Smash")
steve=Avenger("Captain America","punch")
print("*******************************************")
tony.attack()
natasa.attack()
bruce.attack()
steve.attack()
```

```python
class Account:
    count=0
    def __init__(self,name,age,occ,amount,check=False):
        print('no of account holders',Account.count)
        print('===============')
        self.name=name
        self.age=age
        self.occ=occ
        self.amount=amount
        Account.count+=1

    def addmoney(self,newmoney):
        self.amount+=newmoney
        self.check=True
        print('Add money successfully!')
    def withdrawmoney(self,wmoney):
        if self.amount<wmoney:
            print('Not sufficient money')
        else:
            self.amount-=wmoney
            print('success')

    def printdetails(self):
        print('name:',self.name)
        print('Age:',self.age)
        print('Occupation:',self.occ)
        print('Total Amount:',self.amount)
        
p1=Account('Abdul',45,'Service Holder',500000)
p1.addmoney(300000)
p1.printdetails()
print('===============')
p2=Account('rohim',55,'businessman',700000)
p2.withdrawmoney(700000)
p2.printdetails()
print('===============')
p3=Account('Ashraf',62,'officer',200000)
p3.withdrawmoney(250000)
p3.printdetails()
print('===============')
print('no of account holders',Account.count)
```

```python
class Player:
    def __init__(self,name,goalsScored,tacklesWon):
        self.name = name
        self.goalsScored = goalsScored
        self.tacklesWon = tacklesWon
        self.point=0
    def calculatePoint(self):
        self.point+=(self.goalsScored*4)+(self.tacklesWon*3)
        
class Defender(Player):
    def __init__(self,name,goalsScored,tacklesWon,rate):
        super().__init__(name,goalsScored,tacklesWon)
        self.rate=rate
        print('Name:',self.name,',ratting:',self.rate)
    def calculatePoint(self):
        self.rate+=((self.goalsScored*4)+(self.tacklesWon*3)+self.rate)
        print('points of',self.name,'points:',self.rate)
class Attacker(Player):
    def __init__(self,name,goalsScored,tacklesWon,rate):
        super().__init__(name,goalsScored,tacklesWon)
        self.rate=rate
        print('Name:',self.name,',ratting:',self.rate)
    def calculatePoint(self):
        self.rate+=((self.goalsScored*4)+(self.tacklesWon*3)+self.rate)
        print('points of',self.name,'points:',self.rate)
        

print('=========================')
p1 = Defender("Thiago Silva",5,12,8.5)
print('=========================')
p2 = Attacker("Cristiano Ronaldo",14,5,9.0)
print('=========================')
p3 = Attacker("Lionel Messi",12,9,9.5)
print('=========================')
p1.calculatePoint()
print('=========================')
p2.calculatePoint()
print('=========================')
p3.calculatePoint()
```
