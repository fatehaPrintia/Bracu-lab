```python
##question 1
class Calculator:
    def __init__(self):
        print("Let’s Calculate!")
    def add(self,first_value,second_value):
        return first_value+second_value
    def subtract(self,first_value,second_value):
        return first_value-second_value
    def multiply(self,first_value,second_value):
        return first_value*second_value
    def divide(self,first_value,second_value):
        return first_value/second_value

ans=Calculator()
first_value=int(input())
print('value 1:',first_value)
operator=input()
print('operator:',operator)
second_value=int(input())
print('value 2:',second_value)
if operator=='+':
    result=ans.add(first_value,second_value)
if operator=='-':
    result=ans.subtract(first_value,second_value)
if operator=='*':
    result=ans.multiply(first_value,second_value)
if operator=='/':
    result=ans.divide(first_value,second_value)
print("Result:",result)
```

```python
##question 2
class Customer:
    def __init__(self,name):
        self.name=name
    def greet(self,name=None):
        if name!=None:
            print('Hello',name,'!')
        else:
            print('Hello!')
    def purchase(self,*items):
        print(self.name,',you purchased',len(items),'item(s):')
        for i in range(len(items)):
            print(items[i])  
customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")
```

```python
##question 3
class Panda:
    def __init__(self,name,gender,age):
        self.name=name
        self.gender=gender
        self.age=age
    def sleep(self,hour=None):
        if hour==None:
            return f"{self.name}'s duration is unknown thus should have only bamboo leaves"
        elif 3<=hour<=5:
            return f"{self.name}'sleeps {hour} hours daily and should have Mixed Veggies"
        elif 6<=hour<=8:
            return f"{self.name}'sleeps {hour} hours daily and should have Eggplant & Tofu"
        elif 9<=hour<=10:
            return f"{self.name}'sleeps {hour} hours daily and should have Broccoli Chicken"
        
panda1 = Panda("Kunfu","Male", 5)
panda2=Panda("Pan Pan","Female",3)
panda3=Panda("Ming Ming","Female",8)
print("{} is a {} Panda Bear who is {} years old".format(panda1.name,panda1.gender,panda1.age))
print("{} is a {} Panda Bear who is {} years old".format(panda2.name,panda2.gender,panda2.age))
print("{} is a {} Panda Bear who is {} years old".format(panda3.name,panda3.gender,panda3.age))
print('                                                ')
print(panda2.sleep(10))
print(panda1.sleep(4))
print(panda3.sleep())        
```

```python
##question 4
class Cat:
    def __init__ (self,clr=None,act=None ):
        if clr==None and act==None:
            self.clr='White'
            self.act='sitting'
        elif clr!=None and act==None:
            self.clr=clr
            self.act='sitting'
        else:
            self.clr=clr
            self.act=act

    def changeColor(self,newClr):
        self.clr = newClr

    def printCat(self):
        print(self.clr, "cat is", self.act)

c1 = Cat()
c2 = Cat("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
```

```python
##question 5
class Vehicle:
    def __init__(self,x=0,y=0):
        self.x=x
        self.y=y
    def print_position(self):
        print("("+str(self.x)+','+str(self.y)+")")
    def moveUp(self):
        self.y+=1
    def moveDown(self):
        self.y-=1
    
    def moveLeft(self):
        self.x-=1
    def moveRight(self):
        self.x+=1
    
car = Vehicle()
car.print_position()
car.moveUp()
car.print_position()
car.moveLeft()
car.print_position()
car.moveDown()
car.print_position()
car.moveRight()
```

```python
##question 6
class Programmer:
    def __init__(self,name,language,experience,check=False):
        self.name=name
        self.language=language
        self.experience=experience
        self.check=check
    def addExp(self,newexp):
        self.experience+=newexp
        self.check=True
 
    def printDetails(self):
        if self.check==True:
            print('Updating experience of',self.name)
        else:
            print('Horray! A new programmer is born')
 
        print('Name:',self.name)
        print('Language:',self.language)
        print('Experience:',self.experience,'years')       
p1 = Programmer("Ethen Hunt", "Java", 10)
p1.printDetails()
print('--------------------------')
p2 = Programmer("James Bond", "C++", 7)
p2.printDetails()
print('--------------------------')
p3 = Programmer("Jon Snow", "Python", 4)
p3.printDetails()
p3.addExp(5)
p3.printDetails()
```

```python
##question 7
class Student:
    def __init__(self,name,id,department='CSE'):
        self.name=name
        self.id=id
        self.department=department
        print('Name:',self.name)
        print('ID:',self.id)
        print('Department:',self.department)
    def dailyEffort(self,hour):
        self.hour=hour
        print('Daily Effort:',self.hour,'hour(s)')
    def printDetails(self):
        if self.hour<=2:
            print('Suggestion:should give more effort!')
        elif 2<=self.hour<=4:
            print('Suggestion:Keep up the good work!')
        else:
            print('Suggestion: Excellent! Now motivate others.')
            
harry = Student('Harry Potter', 123)
harry.dailyEffort(3)
harry.printDetails()
print('========================')
john = Student("John Wick", 456, "BBA")
john.dailyEffort(2)
john.printDetails()
print('========================')
naruto = Student("Naruto Uzumaki", 777, "Ninja")
naruto.dailyEffort(6)
naruto.printDetails()
```

```python
##question 8
class Patient:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def add_Symptom(self,*symptom):
        self.symptom=symptom
    def printPatientDetail(self):
        print('Name:',self.name)
        print('Age:',self.age)
        print ('Symptoms:',end='')
        for i in self.symptom:
            if i in self.symptom[-1]:
                print(i)
            else:
                print (i,end=',')

p1 = Patient('Thomas', 23)
p1.add_Symptom('Headache')
p2 = Patient('Carol', 20)
p2.add_Symptom('Vomiting', 'Coughing')
p3 = Patient('Mike', 25)
p3.add_Symptom('Fever', 'Headache', 'Coughing')
print("=========================")
p1.printPatientDetail()
print("=========================")
p2.printPatientDetail()
print("=========================")
p3.printPatientDetail()
print("=========================")
```

```python
##question 10
class Shinobi:
    def __init__(self,name,rank,mission=0,salary=0):
        self.name=name
        self.rank=rank
        self.mission=mission
        self.salary=salary
    def changeRank(self,cngRank):
        self.rank=cngRank
        
    def calSalary(self,newmission):
        self.mission+=newmission
        if self.rank=='Genin':
            self.salary=self.mission*50
        elif self.rank=='Chunin':
            self.salary=self.mission*100
        else:
            self.salary=self.mission*500
    def printInfo(self):
        print('Name:',self.name)
        print('Rank :',self.rank)
        print('Number of mission:',self.mission)
        print('Salary:',self.salary)
        
        
naruto = Shinobi("Naruto", "Genin")
naruto.calSalary(5)
naruto.printInfo()
print('====================')
shikamaru = Shinobi('Shikamaru', "Genin")
shikamaru.printInfo()
shikamaru.changeRank("Chunin")
shikamaru.calSalary(10)
shikamaru.printInfo()
print('====================')
neiji = Shinobi("Neiji", "Jonin")
neiji.calSalary(5)
neiji.printInfo()
```
