```python
##TASK 3
class Passenger:
    count=0
    def __init__(self,name):
        self.name=name
        Passenger.count+=1
    def set_bag_weight(self,kg):
        self.__price=0
        self.__kg=kg
        if self.__kg<=20:
            self.__price+=450
        elif 21<=self.__kg<=50:
            self.__price+=450+50
        elif self.__kg>50:
            self.__price+=450+100
    def printDetail(self):
        print('Name:',self.name)
        print('Bus Fare:',self.__price,'Taka')
            
    
print('Total Passenger:', Passenger.count)
p1 = Passenger('Jack')
p1.set_bag_weight(90)
p2 = Passenger('Carol')
p2.set_bag_weight(10)
p3 = Passenger('Mike')
p3.set_bag_weight(25)
print("=========================")
p1.printDetail()
print("=========================")
p2.printDetail()
print("=========================")
p3.printDetail()
print("=========================")
print('Total Passenger:', Passenger.count)
```

```python
##TASK 5
class Fruit:
    fruitCount=0
    def __init__(self,name,count):
        self.name=name
        self.count=count
        Fruit.fruitCount+=count
    @staticmethod
    def saySomethingGood():
        print('Fruits are good for health')
    @staticmethod
    def resetcount():
        Fruit.fruitCount=0
        
apples = Fruit("Apple",3)
pears = Fruit("Pear",4)
print(apples.name, apples.count)
print(pears.name,pears.count)
print("Total number of fruits", Fruit.fruitCount)
Fruit.saySomethingGood()
Fruit.resetcount()
print("Total number of fruits", Fruit.fruitCount)
```

```python
##TASK 6
class Cat:
    Number_of_cats=0
    def __init__(self,clr,action):
        self.clr=clr
        self.action=action
        Cat.Number_of_cats+=1
    @classmethod
    def no_parameter(cls):
        return cls('White','Sitting')
    @classmethod
    def first_parameter(cls,cngclr):
        return cls(cngclr,'Sitting')
    @classmethod
    def second_parameter(cls,cng_action):
        return cls('Gray',cng_action)

    def changeColor(self,updateclr):
        self.clr=updateclr
    def printCat(self):
        print(self.clr ,'is',self.action)
print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
##TASK 7
import math
pi=math.pi
class Cylinder:
    radius=5
    height=18
    def __init__(self,radius,height):
        print('Default radius=', Cylinder.radius ,'and  height=' , Cylinder.height)
        Cylinder.radius=radius
        Cylinder.height=height
        print('Update:Radius=',Cylinder.radius,'and Height=',Cylinder.height)
    @classmethod
    def swap(cls,height,radius):
        obj1=cls(radius,height)
        return obj1
    @classmethod
    def changeFormat(cls,info):
        radius,height=info.split('-')
        obj=cls(float(radius),float(height))
        return obj 
    @staticmethod
    def area(radius,height):
        print('Area:',2*pi*radius*radius+2*pi*radius*height)
    @staticmethod
    def volume(radius,height):
        print('Volume:',pi*radius*radius*height)
c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```
