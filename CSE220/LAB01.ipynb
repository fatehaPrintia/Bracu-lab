```python
##task 1 Shift Left k Cells 

def shiftLeft(source,k):
    for i in range(k):
        for j in range(len(source)-1):
            source[j]=source[j+1]
        source[-1]=0
source=[10,20,30,40,50,60]
shiftLeft(source,3)
print(source)

# ----> output: [40, 50, 60, 0, 0, 0]



##task 2 Rotate Left k cells 
def rotateLeft(source,k):
    for j in range(k):
        t=source[0]
        for i in range(len(source)-1):
            source[i]=source[i+1]
        source[-1]=t
source=[10,20,30,40,50,60]
rotateLeft(source,3)
print(source)

# ----> output:[40, 50, 60, 10, 20, 30]




##task 3 Remove an element from an array

def remove(source,size,idx):
    if idx<0 or idx>size:
        print('wrong index')
        return
    for i in range(idx,size):
        source[i]=source[i+1]
    source[-1]=0

    return source
source=[10,20,30,40,50,0,0]
print(remove(source,5,2))

#  ----> output: [10, 20, 40, 50, 0, 0, 0]



##task 4 Remove all occurrences of a particular element from an array

def  removeAll(source,size,element):
    j=0
    while j<size:
        if source[j]==element:
            for i in range(j,len(source)-1):
                source[i]=source[i+1]
            source[i+1]=0
            
            j-=1
        j+=1
    print(source)
source=[10,2,30,2,50,2,2,60,0,0]
removeAll(source,8,2)

#  ----> output: [10, 30, 50, 60, 0, 0, 0, 0, 0, 0]


##task 5 Splitting an Array

def splitarr(a):
    left_sum=0
    for i in range(len(a)):
        left_sum += a[i]

        right_sum=0
        for j in range(i+1,len(a)):
            right_sum+= a[j]

        if left_sum==right_sum:
            return True
    return False

a= [10, 3, 1, 2, 10]  
print(splitarr(a))
#  ----> output: True



##task 6 arrSeries

def arrseries(n):
    new_arr=[0]*n*n
    size=n*n
    for i in range(1,n+1):
        for j in range(1,n+1):

            new_arr[size-1]=j
            size-=1

        temp=i-1
        while temp>0:
            new_arr[temp+size-1]=0
            temp-=1

    print(new_arr)

arrseries(4)
# ----> output: [0, 0, 0, 1, 0, 0, 2, 1, 0, 3, 2, 1, 4, 3, 2, 1]


###task 7 Max Bunch Count

def maxCount(s):
    start=1
    new_lst=[]
    count1=0
    for p in range(len(s)-1):
        for j in range(p,len(s)-1):
            if s[j]==s[j+1]:
                start+=1
            else:
                new_lst=new_lst+[start]
                start=1
    for i in range(len(new_lst)):
        for k in (i,len(new_lst)-1):
            if new_lst[k]>new_lst[i]:
                max= new_lst[k]
    print(max)
                

s=  [1,1,2, 2, 1, 1,1,1]  
maxCount(s)

#  ----> output: 4



##task 8 repetition
def repetition(s):
    count1=0
    new=[]
    other=[]
    temp=True
    for i in range(len(s)-1):
        if s.count(i)>1:
            new=new+[i]
            
    for j in range(len(new)):
        for k in range(len(s)):
            if new[j]==s[k]:
                count1+=1
        other=other+[count1]
        count1=0

    for i in other:
        for j in range(len(other)):
            if i==other[j]:
                temp=True
            else:
                temp=False
        return temp
s=[4,5,6,6,4,3,6,4]
s1=[3,4,6,3,4,7,4,6,8,6,6]

print(repetition(s))
print(repetition(s1))

#  ----> output:True
#  ----> output:False


##task 1 circular array palindrom

def palindrom(p,start,size):
    boom1=[]
    indx=(start+size-1)
    for i in range(size):
        boom1=boom1+[p[(indx-i)%len(p)]]

    indx=start
    for i in range(size):
        if boom1[i]!=p[(indx+i)%len(p)]:
            return False
    return True

print(palindrom([20,10,0,0,0,10,20,30],5,5)) 
print(palindrom([10,20,0,0,0,10,20,30],5,5))

#  ----> output:True
#  ----> output:False


##task 2 circular array -Intersection

def intersection(a,b,start_1,start_2,size_1,size_2):
    final_arr=[]
    a1=[]
    b1=[]
    indx_1=start_1
    for i in range(size_1):
        a1.append(a[(indx_1+i)%len(a)])
        
    indx_2=start_2
    for j in range(size_2):
        b1.append(b[(indx_2+j)%len(b)])
        
    for k in a1:
        if k in b1:
            final_arr.append(k)
    return final_arr
print(intersection([40,50,0,0,0,10,20,30] , [10,20,5,0,0,0,0,0,5,40,15,25],5,8,5,7))   

#  ----> output: [10, 20, 40]


##task 3 
def musical_chair(chair):
    import random
    while len(chair)!=1:
        num=random.randint(0,4)
        if num == 1:
            rem=len(chair)//2
            for j in range(rem,len(chair)-1):
                chair[j]=chair[j+1]
                j+=1
            chair=chair[:-1]
        else:
            for j in range(len(chair)-1):
                if j==0:
                    te = chair[0]
                chair[j]=chair[j+1]
                j+=1
            chair[-1]=te
    print(chair)
musical_chair(['Fateha','Jannat','Printia','Noumi','Labiba','Disha','Tasfia'])

#  ----> output: ['Jannat']
```
