```python
class Node:
    def __init__(self,value):
        self.value=value
        self.left=self.right=None
        
    def insert(self,val):
        if self.value==None:
            self.value=val
            return 
        if self.value==val:
            return
        if self.value>val:
            if self.left!=None:
                self.left.insert(val)
            else:
                self.left=Node(val)
        else:
            if self.right!=None:
                self.right.insert(val)
            else:
                self.right=Node(val)
        return self
    
    
    def height(self,root1):
        if root1==None:
            return 0
        else:
            return 1+max(self.height(root1.left),self.height(root1.right))
        
        
    def level(self,root1):
        if root1==None:
            return 0
        else:
            return 2+max(self.height(root1.left),self.height(root1.right))
        
#     def level(self):
#         return 1+self.height(root1)



    ##pre order traversal(root,left,right)
    def pre_order(self):
        print(self.value,end=",")
        if self.left!=None:
            self.left.pre_order()
        if self.right!=None:
            self.right.pre_order()
            
    #inorder traversal(left,root,right)
    def in_order(self):
        if self.left!=None:
            self.left.in_order()
        print(self.value,end=",")
        if self.right!=None:
            self.right.in_order()
            
            
    ##postorder traversal(left,right,root)
    def post_order(self):
        if self.left!=None:
            self.left.post_order()
        if self.right!=None:
            self.right.post_order()
        print(self.value,end=",")
        
def is_same(root1,root2):
    if root1==None and root2==None:
        return True
    else:
        if root1!=None and root2!=None:
            return root1.value==root2.value and is_same(root1.left,root2.left) and is_same(root1.right,root2.right)
    return False


def copy(root2):
    print(root2.value,end=",")
    if root2.left!=None:
        copy(root2.left)
    if root2.right!=None:
        copy(root2.right)



root1=Node(10)
lst=[6,3,1,6,98,3,7]
for i in lst:
    root1.insert(i)

print('1)---->The height of the tree is:',root1.height(root1))
print('2)----> The level of the tree is:',root1.level(root1))
print("3)pre order traversal:")
root1.pre_order()
print()
print("4)In order traversal:")
root1.in_order()
print()
print("5)Post order traversal:")
root1.post_order()
###New tree----------------
root2=Node(10)
lst1=[6,3,1,6,98,3,77]
for i in lst1:
    root2.insert(i)
print()
if is_same(root1,root2):
    print("6)The tree's are same")
else:
    print("6)The tree's are different")
# root2.post_order()
# print()
print("Copy of the new tree")
copy(root2)
#----------------------------------------------output-------------------------

# 1)---->The height of the tree is: 4
# 2)----> The level of the tree is: 5
# 3)pre order traversal:
# 10,6,3,1,7,98,
# 4)In order traversal:
# 1,3,6,7,10,98,
# 5)Post order traversal:
# 1,3,7,6,98,10,
# 6)The tree's are different
# 7)Copy of the new tree
# 10,6,3,1,98,77,
```
