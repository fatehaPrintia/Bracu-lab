```python
##task1
class KeyIndex:
    def __init__(self,a):
        self.k=[]
        minn=a[0]
        
        for i in range(len(a)):
            if a[i]<minn:
                minn=a[i]

        self.x=minn*-1      ##making positive
        for i in range(len(a)):
            a[i]+=self.x
            
        maxx=a[0]
        for i in range(len(a)):
            if a[i]>maxx:
                maxx=a[i]

        self.k=[0]*(maxx+1)
        for l in range(len(a)):
            self.k[a[l]]+=1
        
    def search(self,val):
        if val+self.x>=len(self.k):
            return False
        elif  val+self.x<0:
            return False
        elif self.k[val+self.x]==0:
            return False
        else:
            return True
    
    def sorrt(self):
        final=[]
        for m in range(len(self.k)):
            if self.k[m]!=0:
                s=self.k[m]
                count=0
                while count<s:
                    final=final+[m-self.x]
#                     print(m-self.x,end=",")
                    count+=1
                m+=1
            else:
                m+=1
        return final

a=[-1,3,-2,4,6]
b=KeyIndex(a)
print(b.search(-1))
b.sorrt()
# -----------------output-------------------------------
# True
# [-2, -1, 3, 4, 6]



##task2

def hashh(a):
    TABLE=[0]*9
    for i in range(len(a)):
        count=0
        summ=0
        for char in a[i]:
            if char >="A" and char<="Z":
                if char not in "AEIOU":
                    count+=1
            elif char>="0" and char<="9":
                summ+=int(char)
        

        final=(count*24+summ)%9
        x=(final)%len(TABLE)

        if TABLE[x]==0:
            TABLE[x]=a[i]

        elif TABLE[x]!=0:
            while TABLE[x]!=0:
                x+=1
                if x==len(TABLE):
                    x=0
                
            TABLE[x]=a[i]

    print(TABLE)
        
        
hashh(['ST1E89B8A32','ST1E89B8A33','ST1E89B8A34','ST1E89B8A30','ST1E89B8A36','ST1E89B8A27'])
# ----------------------output----------------------------------
# ['ST1E89B8A27', 0, 'ST1E89B8A30', 0, 'ST1E89B8A32', 'ST1E89B8A33', 'ST1E89B8A34', 0, 'ST1E89B8A36']
```
