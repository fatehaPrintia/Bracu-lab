# Linear to circular array convert

```python
##convert the linear array to circular array starting at 6th index
def converter(source,start,size):
    circular_array=[0]*len(source)
    for i in range(size):
        circular_array[start]=source[i]
        start=(start+1)%len(circular_array)
        i+=1
    return circular_array
source=[10,20,30,40,50,0,0,0,0,0,0]
print(converter(source,8,5))
```

```python
#converting a circular array to a linear array

def circularize(linear,start,size):
    circular=[0]*len(linear)
    index=start
    i=0
    while(i<size):
        circular[index]=linear[i]
        index=(index+1)%len(circular) 
        i=i+1
    return circular
linear=[10,20,30,40,50,0,0,0,0,0,0]
circ=circularize(linear,8,5)
print(circ)
```

# Backword/reverse printing a linear array

```python
def reverse_linear(a,size):
    idx=size-1
    for i in range(size):
        print(a[idx])
        idx-=1
a=[10,20,30,40,50,0,0,0,0]
reverse_linear(a,5)
```

# Backword/reverse printing a circular array

```python
#  reverse printing 
def reverseprint(a,start,size):
    index=(start+size-1)%len(a)
    for i in range(size):
        print(a[index])
        index-=1
        if index<0:
            index=len(a)-1
circularArray=[40,50,0,0,0,0,0,0,10,20,30] 
reverseprint(circularArray,8,5)
```

```python
##resize Circular array

def resize_circular(source,start,size):
    new_array=[0]*(len(source)+2)
    indx_cir=start
    indx_new=start
    for i in range(len(source)):
        new_array[indx_new]=source[indx_cir]
        indx_cir=(indx_cir+1)%len(source)
        indx_new=(indx_new+1)%len(new_array)
        i+=1
    return new_array
source=[20,30,40,50,10]
resize_circular(source,4,5)
    
```

```python
#resizing a circular array
def resize(circ,start,size):
    resizedCirc=[0]*(len(circ)+2)
    index_circ=start
    index_resizedCirc=start
    i=0
    while(i<size):
        resizedCirc[index_resizedCirc]=circ[index_circ]

        index_circ=(index_circ+1)%len(circ)

        index_resizedCirc=(index_resizedCirc+1)%len(resizedCirc)  

        i=i+1
    return resizedCirc
circ=[20,30,40,50,10] #creating a circular array with start 4 and size 5
circ=resize(circ,4,5) 
print(circ)
```

# Inserting an element at the end of circular array

```python
def insert_circular(array,start,size,item):
    circular=[0]*(len(array)+2)
    indx_array=start
    indx_cir=start
    for i in range(size):
        circular[indx_cir]=array[indx_array]
        indx_array=(indx_array+1)%len(array)
        indx_cir=(indx_cir+1)%len(circular)
        i+=1
    indx1=(start+size)%len(circular)
    circular[indx1]=item
    size+=1
    return circular

array=[4,5,1,2,3]
insert_circular(array,2,5,6)
```

# Deleting an element from the end

```python
def del_element(cir,size,start):
    idx=start
    x=[]
    for i in range(size):
        x=x+[cir[idx]]
        idx+=1
        if idx==len(cir):
            idx=0
    for j in range(len(x)):
        if j==len(x)-1:
            x[-1]=0
    print(x)
arr=[40,50,0,0,0,0,0,0,10,20,30]
del_element(arr,5,8)
```
