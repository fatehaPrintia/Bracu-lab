```python
class Node:
    def __init__(self,element,next,prev):
        self.element=element
        self.next=next
        self.prev=prev


# Task 1 (ii)
class Dummy_Headed_Doubly_Circular_Linked_list:
    def __init__(self,a):
        self.head=Node(None,None,None)


#Task 2

class Node:
    def __init__(self,element,next,prev):
        self.element=element
        self.next=next
        self.prev=prev

class Dummy_Headed_Doubly_Circular_Linked_list:
    def __init__(self,a):
        self.head=Node(None,None,None)
        dummy_Node=None
        for i in a:
            if len(a)==0:
                print("The list is empty")
            else:
                val=Node(i,None,None)
                if dummy_Node==None:
                    self.head.next=self.head.prev=val
                    val.next=val.prev=self.head                       #task 2(1(a))
                    dummy_Node=val
                else:
                    dummy_Node=self.head.prev
                    self.head.prev=dummy_Node.next=val
                    val.prev=dummy_Node
                    val.next=self.head
                
    def showList(self):
        n=self.head.next
        if n==None:
            print('The list is empty')
        else:
            while n!=self.head:
                print(n.element,end=",")                              #task 2(2)
                n=n.next


    def insert(self,value):
        n=self.head.next
        val=Node(value,None,None)
        count=False
        if n==None:
            return " Can not be remove as the list is empty"            #task 2(3)
            
        while n.next!=self.head:
            if n.element==value:
                count=True
                break
            n=n.next
        if count==True:
            print('Already Exist')
            return
        else:
            n.next=val
            val.prev=n
            val.next=self.head
            self.head.prev=val
        
    def insert_indx(self,value,indx):
        val=Node(value,None,None)
        n=self.head.next
        count=0
        temp=False
        if n==None:                                                    #task 2(4)
            return " Can not be remove as the list is empty"
            
        while n!=self.head:
            if value==n.element:
                temp=True
                break
            n=n.next
            count+=1
            
        if temp==True:
            print("Already Exist")
            return
        elif count<indx:
            print("Invalid Index")
            return
        elif n==None:
            print("empty")
            
        else:
            n=self.head.next
            count=0
            while n!=self.head:
                if count==indx:
                    previous_node=n.prev
                    previous_node.next=val
                    n.prev=val
                    val.next=n
                    val.prev=previous_node

                n=n.next
                count+=1
            return n
            
            
    def remove(self,indx):
        n=self.head.next
        count=0                                                                      #task 2(5)
        if n==None:
            return " Can not be remove as the list is empty"
            
        while n!=self.head:
            count+=1
            n=n.next
        if count<indx:
            print("Invalid index")
            return

        else:
            n=self.head.next
            count=0
            while n!=self.head:
                if count==indx:
                    p=n.prev
                    q=n.next
                    p.next=q
                    q.prev=p
                count+=1
                n=n.next
            return n
            
    def removeKey(self,deletekey):
        n=self.head.next
        if n==None:
            return " Can not be remove as the list is empty"
        else:
            n=self.head.next
            while n!=self.head:
                if n.element==deletekey:                                               #task 2(6)
                    q=n.next
                    p=n.prev
                    p.next=q
                    q.prev=p
                    n.prev=n.next=None
                    return n.element
                n=n.next

#------------------------------------------Tester-------------------------------------------
a=[10,20,30,40,50]
lst=Dummy_Headed_Doubly_Circular_Linked_list(a)
lst.insert(60)
print("insert in tail:")
lst.showList()
lst.insert_indx(90,2)
print("\n")
print("insert at index:")
lst.showList()
print("\n")
lst.remove(3)
print("\n")
print("remove a index:")
lst.showList()
print("\n")
print('Deletedkey:',lst.removeKey(40))
print("\n")
print("remove value:")
lst.showList()

#------------------------outputs--------------------------------------------------------
# insert in tail:
# 10,20,30,40,50,60,

# insert at index:
# 10,20,90,30,40,50,60,



# remove a index:
# 10,20,90,40,50,60,

# Deletedkey: 40


# remove value:
# 10,20,90,50,60,
```
